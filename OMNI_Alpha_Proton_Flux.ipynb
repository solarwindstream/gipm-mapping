```python
#import OMNI alpha/proton data and process averages for all relevant windows.

#making function: OMNI_cdf_conv_ap_ratio(filename)

#OMNI_cdf_path = '/Users/roseatkinson/Documents/omni_hro2s_1min_20010201000000_20240630000000_cdaweb.cdf'
#OMNI_NaNp_df = OMNI_cdf_conv_ap_ratio(OMNI_cdf_path)

#OMNI_csv_path = '/Users/roseatkinson/Documents/OMNI_23yrs_NaNp.csv'
#OMNI_NaNp_df.to_csv(OMNI_csv_path)

omni_df = pd.read_csv('/Users/roseatkinson/Documents/OMNI_23yrs_NaNp.csv',encoding='utf-8')
omni_df['datetime'] = pd.to_datetime(omni_df['datetime'], format='mixed')
omni_df.set_index('datetime', inplace = True)
```

```python
#import Cluster integrated CSVs
##load CSVs

cl_file_list = []

path = "/Users/roseatkinson/Documents/Cluster_Integrated_CSVs/**"

for path in glob.glob(path, recursive=True):
    if '.csv' in path:
        cl_file_list.append(path)
```

```python
#just go csv by csv, adding the new column!
loop_counter = 0

for file in cl_file_list:
    df = pd.read_csv(file,encoding='utf-8')
    if 'Na_Np (mean)' in df.columns:
        loop_counter = loop_counter + 1 
    else:
        df['datetime'] = pd.to_datetime(df['datetime'], format='mixed')
        windows_list = df['datetime'].values.tolist()
        df.set_index('datetime', inplace = True)
        om_averages = omni_seg_ap_ratio(omni_df, windows_list)
        joined_df = df.join(om_averages)
        joined_df.to_csv(file)
        loop_counter = loop_counter + 1 
        
    if loop_counter % 1000 == 0:
        print(loop_counter)
```
